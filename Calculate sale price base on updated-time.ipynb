```python
sales= pd.read_csv('sales.csv')
print(sales.head(15))
```

```python
sales.info()
prices.info()
```

```python
# change data type of dataframe
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'])
prices['updated_at'] = pd.to_datetime(prices['updated_at'])
```

```python
sales.info()
prices.info()
```

```python
sales= sales.sort_values('ordered_at')
prices=prices.sort_values('updated_at',)
```

```python
sale_price=pd.merge_asof(sales, prices,
                         by='product_id',
                         left_on='ordered_at',
                         right_on='updated_at',
                         direction='nearest')
sale_price
```

```python
#  Create condition
conditions=[sale_price['ordered_at']>= sale_price['updated_at'],
           sale_price['ordered_at']< sale_price['updated_at']]
results=[sale_price['new_price'],sale_price['old_price']]

# create new column called final_price which is price that customer actually had to pay when they ordered

sale_price['final_price']=np.select(conditions, results)
sale_price
```

```python
# create new column called "revenue" which calculate revenue of each product base on time that product had been ordered

sale_price['revenue']= sale_price['final_price']*sale_price['quantity_ordered']
sale_price
```

```python
# Calculate  total revenue by product_id
sale_price.groupby(['product_id']).agg({'revenue':'sum'})
```

```python
# Calculate  revenue by product_id and final_price
sale_price.groupby(['product_id','final_price']).agg({'revenue':'sum'})
```
